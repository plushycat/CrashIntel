```python
import pandas as pd

# Loads the dataset
file_name = r"Datasets\integrated_traffic_dataset (1).csv"
df = pd.read_csv(file_name)

# 1. Total duplicate rows (across all columns)
total_duplicates = df.duplicated().sum()
print(f"Total duplicates found: {total_duplicates}")

# 2. Shows first 5 duplicate rows 
if total_duplicates > 0:
    duplicates = df[df.duplicated()].head()
    print("First 5 duplicate rows:\n", duplicates)
else:
    print("No duplicate rows to display.")

# 3. Removes duplicates (keep first occurrence)
df_no_dupes = df.drop_duplicates()

output_file_name = "integrated_traffic_dataset_no_duplicates.xlsx"
df_no_dupes.to_excel(output_file_name, index=False)

print(f"Cleaned dataset saved as {output_file_name}")
print(f"Original rows: {len(df)}")
print(f"Rows after removing duplicates: {len(df_no_dupes)}")
```
