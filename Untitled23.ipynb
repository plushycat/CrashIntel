```python
import pandas as pd

# Load dataset (Excel file)
df = pd.read_excel(r"C:\Users\BHOOMIKA_B_G\OneDrive\Desktop\integrated_traffic_dataset (1).csv.xlsx")

# ðŸ”Ž 1. Total duplicate rows (across all columns)
print("Total duplicates:", df.duplicated().sum())

# ðŸ”Ž 2. Show duplicate rows
duplicates = df[df.duplicated()]
print("Duplicate rows:\n", duplicates)

# ðŸ”Ž 3. Remove duplicates (keep first occurrence)
df_no_dupes = df.drop_duplicates()

# Save cleaned dataset
df_no_dupes.to_excel(r"C:\Users\BHOOMIKA_B_G\OneDrive\Desktop\integrated_traffic_dataset_no_duplicates.xlsx", index=False)

print(" Cleaned dataset saved as integrated_traffic_dataset_no_duplicates.xlsx")
```
